### Alternate decompositions

One advantage of the Fourier decomposition is its orthogonality. 
Imagine if there is a function composed of two modes: 

$$
f(x) = cos (2 \pi x) + cos(4 \pi x)
$$
defined over the interval 0 to 1. 

Then the variance of the full signal is equal to the variance of the parts. It can be shown that similarly if we decompose a signal into its fourier components then the total variance is equal to the sum of the variance of the parts. (The Plancheral relationship). 

```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
x = np.arange(0,1,0.01)

f_l = np.cos(2*np.pi*x)
f_s = np.cos(8*np.pi*x)
f = f_l + f_s
```

```python
plt.plot(x, f_l)
plt.plot(x, f_s)
plt.plot(x, f)
```

```python
np.mean(f**2), np.mean(f_l**2), np.mean(f_s**2), np.mean(f_l**2)+np.mean(f_s**2) 
```

However if we use a filter, then this is not necessarily true. That is if we use a filter to decompose :
$$
f = \overline{f}^l + f^a
$$
where $f^a = f - \overline{f}^l$ is the anomaly field. Then
$<|f|^2> = <|\overline{f}^l|^2> + <|f^a|^2> +  2 <|\overline{f}^l f^a|>$.
The correlation between the full and anomaly field is not generally zero. 

This is the reason that some studies on filtering define the variance or KE of the large scale field as $<|\overline{f}^l|^2>$ but the variance of the small scale field as $<\overline{|f|^2}^l> - <|\overline{f}^l|^2>$.



**We had defined small scale earlier and looked at the variance of that as a function of scale.**
I think that can be an alternative definition. Under this definition also we have a kind of cumulative spectrum and we can look at its variation with scale. 

Roughly speaking: 

$$
u(x) = \int \hat{u}(k) e^{ikx} dk
$$
$$
u_l(x) = \int  \hat{G}(k) \hat{u}(k) e^{ikx} dk
$$
So:
$$
u_a(x) = u(x) - u_l(x) = \int  [1- \hat{G}(k)] \hat{u}(k) e^{ikx} dk
$$

We can see that
$$
E(k) \neq E_l (k) + E_a(k)  = |\hat{G}|^2 E(k) + |1-\hat{G}|^2 E(k)
$$
and the term on the far RHS is not simply E(k). 

We can still define a quantity:
$$
\epsilon_a (l) = u_a^2,
$$
similar to the cumulative spectrum in Sadek and Aluie. This goes to zero at small filter scales and approaches the total variance at large filter scales (like the reverse of the cumulative spectrum defined in the paper). One could probably work through the math and show how the two are related. I think in the presence of some large scale mean, things might get a bit tricky.
