## Filtering spectrum 

**Sadek and Aluie 2018** defined a filtering spectrum.

For a filtered field 
$$
u_l(x) = \int u(y) G_l (x - y) dy
$$

The cumulative spectrum is defined as:
$$
\epsilon(k_l) = 1/2<|u_l(x)|^2>,$$
where $k_l=L/l$.
Notice that as filter scale becomes really small ($k_l$ is large), we recover the original signal. While when filter scale becomes very large ($k_l$ approaches 0) we recover the signal mean. [This definition is apparently also available in Uriel Frisch's book on turbulence].

and the Filtering spectrum is:
$$
\overline{E}(k_l) = \frac{d}{d k_l} \epsilon(k_l) 
$$

> Note that this filtering spectrum is capable of obtaining the same power law scaling as the true spectrum, but should not be expected to match the true specturm in an absolute sense. See appendix in Sadek and Aluie 2018 to appreciate this better.
*Probably if we see a match, it is just coincidence or approximate*.

**Steinberg et al 2019** have a slightly different explanation of the method, but basically the same idea. 

```python
import numpy as np
import scipy.fft as fft
import matplotlib.pyplot as plt
import funcs
```

```python
L = 1000 # let's say units of m 
N = 512

alpha = 2
f, x, dx = funcs.random_phase_power_law_function(alpha, N, L)

f_ps, k = funcs.power_spectrum(f, dx)
```

```python
plt.figure(figsize=(12,5))

plt.subplot(121)
plt.plot(x, f)

plt.subplot(122)
plt.plot(k, f_ps, 'k.-')
plt.plot(k, 1e-5*k**-alpha, '--')
plt.xscale('log')
plt.yscale('log')

#plt.ylim([5e-3, 2])
```

#### Note on length scales 

Annoyingly we need to think about this differently with every paper because notation changes. In Sadek and Aluie the exponential in def of Fourier transform is chosen to be 
$$
e^{i \frac{2 \pi}{L} k x}
$$
Here k is an integer. *This is because of periodic domain, and the condition that $e^{i \frac{2 \pi}{L} k x} = e^{i \frac{2 \pi}{L} k (x+L)}$ is only satisfied for integer k.*

The interpretation of this integer wavenumber k can be $L/l$. where for k=1 we have that $l=L$, for k=2 we have that $l=L/2$ and so on. So we say that k is associated with length scale l. 

*Note that this discreteness is a result of periodicity, and would not be essential in the CTFT (but that only works in continuous and infinite domains).*

> So when we pick length scales to filter, we **may** pick them as 
$L/k$. So we will get $l = Ndx, Ndx/2, Ndx/3, ... , Ndx/(N/2-1), 2dx$. When we pick the filter scale in terms of number of grid points, we can drop the dx (or set to 1). 

However, since there is not a one to one match between wavelength and filter scale, the above statement is a bit arbitrary. Said otherwise a filter scale set as the standard deviation of a Gaussian kernel does not have an exact correspondence to a wavelength, just an approximate one can be decided on. 
Also in some discrete filters (like top hat) there is an amibiguity to filter width, and so many choices of l will be the same. 



```python
#l_n = N/ np.arange(1, N/2+1) 
#l = np.arange(1, N/2) 

l = 2*dx*np.arange(1,3*N/4) 
# we put factor of two here because top-hat function can't be defined at odd number
# of grid points.

l_n = l/dx # number of grid points for length scale

m = L / l  # L/l this is wave number as used in Sadek and Aluie 2018
k_l = m / L # This is defined to match how k is defined in fft (https://numpy.org/doc/stable/reference/generated/numpy.fft.fftfreq.html)
# we define k_l this way so we can use same axis as what we used to make our data. 
```

```python
f_l_top = np.zeros((N, len(l)))
f_l_gauss = np.zeros((N, len(l)))
f_l_vdv1 = np.zeros((N, len(l)))
f_l_vdv2 = np.zeros((N, len(l)))


for i in range(len(l)):
    f_l_top[:,i]    =  funcs.my_filter1d(f,  l_n[i], funcs.top_hat_kernel)
    
    f_l_gauss[:,i]  =  funcs.my_filter1d(f, l_n[i]/np.sqrt(12), funcs.gaussian_kernel)
    
    f_l_vdv1[:,i] = funcs.my_filter1d_vdv(f, l_n[i]/np.sqrt(12), funcs.G_m,  m=1, mode='wrap', truncate=8.)
    f_l_vdv2[:,i] = funcs.my_filter1d_vdv(f, l_n[i]/np.sqrt(12), funcs.G_m,  m=2, mode='wrap', truncate=8.)
    
```

```python
plt.figure(figsize=(10,9))

plt.subplot(221)
plt.pcolormesh(l, x, f_l_top)
plt.xscale('log')
plt.ylabel('x')
plt.xlabel('l')
plt.title('Top hat')

#plt.xlim([1, 500])

plt.subplot(222)
plt.pcolormesh(l, x, f_l_gauss)
plt.xscale('log')
plt.ylabel('x')
plt.xlabel('l')
plt.title('Gaussian')

plt.subplot(223)
plt.pcolormesh(l, x, f_l_vdv1)
plt.xscale('log')
plt.ylabel('x')
plt.xlabel('l')
plt.title('VDV m =1')

plt.subplot(224)
plt.pcolormesh(l, x, f_l_vdv2)
plt.xscale('log')
plt.ylabel('x')
plt.xlabel('l')
plt.title('VDV m =2')

plt.tight_layout()
```

As a reminder the cumulative spectrum is defined as:
$$
\epsilon(l) = \epsilon(k_l) = 1/2<|u_l(x)|^2>
$$
where the functional dependence on $l$ or $k_l$ just depends on how we make the plot. 

```python
# cumulative spec
cum_spec_top = 0.5 * np.mean(f_l_top**2, axis=0)
cum_spec_gauss = 0.5 * np.mean(f_l_gauss**2, axis=0)
cum_spec_vdv1 = 0.5 * np.mean(f_l_vdv1**2, axis=0)
cum_spec_vdv2 = 0.5 * np.mean(f_l_vdv2**2, axis=0)
var_f = 0.5 * np.mean(f**2)
```

```python
var_f, np.sum(f_ps)
```

```python
plt.figure(figsize=(10, 4))
plt.subplot(121)
plt.plot(l,  cum_spec_gauss, '.')
plt.plot(l,  cum_spec_top, '.-')
plt.plot(l,  cum_spec_vdv1, '-')
plt.plot(l,  cum_spec_vdv2, '-')
plt.xscale('log')
plt.xlabel('l')

plt.subplot(122)
plt.plot(k_l, cum_spec_gauss, '.')
plt.plot(k_l, cum_spec_top, '-')
#plt.plot(k_l,  cum_spec_vdv1, '-')
plt.plot(k_l,  cum_spec_vdv2, '-')
plt.hlines(var_f, np.min(k_l), np.max(k_l), linestyles='--')
plt.xscale('log')
plt.xlabel('k')
```

... and the Filtering spectrum is:
$$
\overline{E}(k_l) = \frac{d}{d k_l} \epsilon(k_l) 
$$
A measure of variance per unit wavenumber.

```python
#k_l = 1/l*N
k_mid = 0.5*(k_l[1:] + k_l[:-1])
dk = np.mean(np.diff(k))
dk_l = (k_l[1:] - k_l[:-1])

filt_spec_gauss_k = (cum_spec_gauss[1:] - cum_spec_gauss[:-1])/ dk_l 
filt_spec_top_k   = (cum_spec_top[1:] - cum_spec_top[:-1])/ dk_l 
filt_spec_vdv1_k  = (cum_spec_vdv1[1:] - cum_spec_vdv1[:-1])/ dk_l 
filt_spec_vdv2_k  = (cum_spec_vdv2[1:] - cum_spec_vdv2[:-1])/ dk_l 
```

```python
# The "integrals" of the filtered spectra do approach the variance.
np.sum( filt_spec_gauss_k * np.abs(dk_l)  ), np.sum( filt_spec_vdv2_k * np.abs(dk_l)  ) 
```

```python
Ek = f_ps / dk # this gives us the energy per unit wavenumber. 
```

Why did we divide by dk above? 
- We did this because we want the discrete integrals of the true spectrum and the filtering spectrum to match. However, the discrete grid used by our filtering is stretched out at larger k. So we need to look at the per unit k functions, but the f_ps from fft sums to the total variance (rather than integrates, but it doesn't matter as dk = 1 in that case). Think of just as a fixing the units of f_ps to match those of filtering_spectrum, since we changed the units on k axis to be in real units rather than integers. 

```python
np.sum(Ek * dk), np.sum( filt_spec_vdv2_k * np.abs(dk_l)  ) # some difference because we don't have full range of scales. 
```

```python
plt.figure(figsize=(6, 4))

plt.plot(k_mid, filt_spec_top_k, '.-', label='Top hat kern')
plt.plot(k_mid, filt_spec_gauss_k, '.-', label='Gauss kern')
plt.plot(k_mid, filt_spec_vdv1_k, '.-', label='vdv1')
plt.plot(k_mid, filt_spec_vdv2_k, '.-', label='vdv2')

plt.plot(k, Ek, 'k--', label='500XTrue spectrum')
#plt.hlines(var_f, )
plt.xscale('log')
plt.vlines(1/2/dx, 1, 100, label='Nyquist freq')
plt.ylim([0.01, 2000])
plt.yscale('log')
plt.xlabel('k')
plt.legend()
```

```python
plt.figure(figsize=(6, 4))

plt.plot(k_mid, filt_spec_top_k, '.-', label='Top hat kern')
plt.plot(k_mid, filt_spec_gauss_k, '.-', label='Gauss kern')
plt.plot(k_mid, filt_spec_vdv1_k, '.-', label='vdv1')
plt.plot(k_mid, filt_spec_vdv2_k, '.-', label='vdv2')

plt.plot(k, Ek, 'k--', label='500XTrue spectrum')
#plt.hlines(var_f, )
#plt.xscale('log')
plt.vlines(1/2/dx, 1, 100, label='Nyquist freq')
plt.ylim([0.01, 2000])
plt.xlim([0, 0.04])
#plt.yscale('log')
plt.xlabel('k')
plt.legend()
```

So we learn a few things: 
- There is no exact analogy between the filter length scales and wavenumbers, which is why the filtering spectrum seems shifted in k. 
- Also there is no exact match in the value of the filtering and true spectrum. The integral of both of them over k is the variance of the signal. 
     - there might be better match if there was less stretching to higher k near the place where the slope changes.
     - The integral over k of both does match. 
     - The scaling relationships are also similar in regions with power law behavior. 
- We can matched the filter scale from top hat with that of Gaussian by setting the standard deviation of the two filters to be the same. This leads to a gaussian filter scale to be the top hat filter scale divided by sqrt(12). 
- As shown in figure below, it is important to properly place the derivative at the mid-point. 
- If you make the spectrum too steep then only vdv m=2 filter captures the right scaling. (**Check by changing the alpha and re running notebook**).
- Because of the need to define the filtering scales in discrete sense we have very few points at large k, which is the revese of fft - where most of the points are conc near large k.
