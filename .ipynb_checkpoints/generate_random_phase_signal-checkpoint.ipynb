This notebook goes through a detailed explanation of how random phase signals can be generated with a given spectrum. 

We will need the DFT relations (https://docs.scipy.org/doc/scipy/tutorial/fft.html#d-discrete-fourier-transforms)

$$
G[k] = \sum_{n=0}^{N-1} f[n] e^{-2 \pi i \frac{kn}{N}}
$$

$$
f[n] = \frac{1}{N}\sum_{k=0}^{N-1} G[k] e^{2 \pi i \frac{kn}{N}}
$$


and the corresponding Plancharel identity is:
$$
\sum_{n=0}^{N-1} |f[n]|^2 = \frac{1}{N} \sum_{k=0}^{N-1} |G[k]|^2
$$

```python
import numpy as np
import scipy.fft as fft
import matplotlib.pyplot as plt
```

```python
L = 2*np.pi
N = 128 # Change N to change resolution of the signal
dx = L/N

n= np.arange(0, N)
x = n*dx
```

```python
k = np.arange(0, N)
```

It is useful to work with statistical averages like mean and variance. The variance (*different from traditional definition, where the mean is removed*) or energy is defined as (and the equality follows from Plancheral's theorem): 
$$
\frac{1}{N} \sum_{n=0}^{N-1} |f[n]|^2 = \frac{1}{N^2} \sum_{k=0}^{N-1} |G[k]|^2 = \sum_{k=0}^{N-1} |F[k]|^2,
$$
where $F[k] = G[k]/N$ is defined to make things appear more convenient, even though internally fft uses $G[k]$. It is also customary to add a factor of 1/2. 
The above equation shows that the variance can be decomposed into sums of Fourier coefficients. 

Before moving forward we need to understand a few more related key ideas. 

#### Length scales 
Just for this discussion we use m to represent the integer wavenumber (what was called k in the above text) and k to represent the wavenumber with units (1/[m] for length and 1/[s] for time)
$$
e^{i 2 \pi k x} = e^{i 2 \pi (\frac{m}{Ndx}) ndx}
$$
so the k with units corresponding to a particular unitless integer m is $\frac{m}{Ndx}$ (indicating number of waves per L $=Ndx$). So an estimate of the the length scale corresponding to some m is $[\frac{Ndx}]{m}$. 

>*Notice that here the $2 \pi$ has not been absorbed into the wavenumber, so k is not the angular wavenumber (units radians/m)*


```python
# We have two ways to figure out the length scales
# a) We can use the formula we have above directly 
# (but need to take care of how the freq are arranged)
# b) use the fftfreq function 

inc_k = slice(1, int(N/2)) # remember that this will be from 1 - N/2-1 according to python convention
lscales = N*dx/k[inc_k]

fft_freq = fft.fftfreq(N, d = dx) # lscales will be inverse of this
```

```python
print( max(fft_freq), min(fft_freq), 1/2/dx) 
# note that the nyquist frequency with odd number of data points is 
# only stored with the negative ones.
```

```python
plt.plot(k, 1./fft_freq)
plt.plot(k[inc_k], lscales, '.')

plt.xscale('log')
```

```python
ngrid, kgrid = np.meshgrid(n, k)
#waves = 

waves = np.cos(2*np.pi *kgrid * ngrid/N)
```

As k goes from 0 to N-1, the size of the waves decreases (k=0 is the mean, k=1 is the longest wave possible in the periodic domain, and so is k=N-1) and then increases because the negative frequencies are present. 

```python
plt.plot(x, waves[:,1])
plt.plot(x, waves[:,int(N/4)])
plt.plot(x, waves[:,int(N/2)], '.')
plt.plot(x, waves[:,int(3*N/4)])
plt.plot(x, waves[:,-1])
```

#### Energy spectrum 

We want to create signals with preseribed energy spectra. 

The energy spectra is defined as: 
$$
\langle \frac{1}{2} |f(x)|^2 \rangle = \sum_{\mathbf{k}} \frac{1}{2} |F[k]|^2 = \sum_{k=0}^{\infty} E(k)
$$

where for the first equality on the RHS there are separate contributions form the positive and negative frequencies, but for the second equality the spectrum contains both of them. This means that $E(k) = \frac{1}{2} (|F[k]|^2 +  |F[-k]|^2)$. 

Also since we are dealing with real functions only, the Fourier coefficients are symmetric : $F[k] = F[-k]$. So $E(k) = |F[k]|^2$. 

This above result also means that we can write everything as a Fourier series :
Using the def of Fourier sums
$$
f[n] = \frac{1}{N}\sum_{k=0}^{N-1} G[k] e^{2 \pi i \frac{kn}{N}}
$$
we get
$$
f[n] = \sum_{k=0}^{N/2} 2 F[k] \cos (2 \pi i \frac{kn}{N})
$$
as the $\sin$ terms drop out. 
> Some care must be taken about the limits of the sum, since for even N the positive frequencies end on $N/2-1$ and $N/2$ on negative side, and foe odd N the positive and negative frequencies extend to $(N-1)/2$. 



##### Random phases 

We create random phase signal by choosing 
$F [k] = |F[k]|e^{i\theta} = E[k]^{1/2} e^{i\theta}$, where $\theta$ is a randomly chosen angle in radians (between 0 and $2\pi$). 
This works because the $|e^{i \theta}| = 1$.



```python
k_positive = k[inc_k] # The case for even
k_positive_units = k_positive/ N /dx
```

```python
k_smallest = k_positive_units[0]
k_largest = k_positive_units[-1]

k_break = k_smallest*5 # where slope break is

fhat_random_phase =np.zeros_like(k).astype(np.complex128)

Ek = np.zeros_like(k_positive_units)

alpha = 2. # power law slope

C = 1/ (k_break**(-alpha)) # matching coeff.

Ek[k_positive_units<=k_break] = 1.
Ek[k_positive_units>k_break] = C*k_positive_units[k_positive_units>k_break]**(-alpha) # Note that the energy spectrum is square of fhat

random_theta = np.random.uniform(low=0, high=2*np.pi, size=(len(Ek),))

fhat_random_phase[inc_k] = Ek**0.5 * np.exp(1j*random_theta)

# Set negative freq to be the same
fhat_random_phase[-1: - int(N/2): -1] = np.conjugate(fhat_random_phase[inc_k])
# We have ended up with the nyquit freq set to 0. 
```

```python
plt.plot(k_positive_units, Ek, '.-')
plt.plot(fft_freq, np.abs(fhat_random_phase)**2, '.')
plt.plot(k_positive_units, k_positive_units ** -alpha, '--')

plt.xscale('log')
plt.yscale('log')
```

```python
plt.plot(fft_freq, np.abs(fhat_random_phase), '.-')
#plt.xlim([-4,4])
#plt.yscale('log')
```

```python
fhat_random_phase[1], fhat_random_phase[-1]
# notice that these are complex conjugates
```

##### Generate signals 

There are two ways to generate the signals:
- Use the ifft function
- Sum manually with the help of cosines. $
f[n] = \sum_{k=0}^{N/2} 2 |E[k]|^{0.5} \cos (2 \pi i \frac{kn}{N} + \theta)
$

```python
f_gen1 = fft.ifft(N*fhat_random_phase)
```

```python
plt.plot(x, f_gen1.real)
plt.plot(x, f_gen1.imag)
# Since we created the fhat's very carefully, the imaginary part is ~ 0.
```

```python
plt.plot(x, f_gen1.imag)
# note the scale on the figure. 
```

```python
ngrid, kgrid = np.meshgrid(n, k_positive)
ngrid, thetagrid = np.meshgrid(n, random_theta)
ngrid, Ekgrid = np.meshgrid(n, Ek)
```

```python
ngrid.shape
```

```python
f_gen2 = np.sum(2 * Ekgrid**0.5 * np.cos(2*np.pi*kgrid*ngrid/N + thetagrid), axis=0)
```

```python
plt.plot(x, f_gen1.real)
plt.plot(x, f_gen2, '.')
# The two methods give the exact same signals (as they should! )
```

```python
plt.plot(x, f_gen1.real - f_gen2)
```

The Plancharel's relationship is satisfied perfectly: 

```python
np.sum(Ek), np.mean( 0.5* f_gen1.real** 2), np.mean( 0.5* f_gen2** 2)
```

#### Confusion about units
https://dsp.stackexchange.com/questions/58270/difference-between-ctft-and-dtft

The units of the DFT is different from those of the CTFT, but this is due to a sampling dependent scaling factor. Since the scaling factor is a constant with units for a particular case, it should (hopefully) not influence power law behaviours.
