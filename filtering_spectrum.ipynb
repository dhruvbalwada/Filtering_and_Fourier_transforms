## Filtering spectrum 

**Sadek and Aluie 2018** defined a filtering spectrum.

For a filtered field 
$$
u_l(x) = \int u(y) G_l (x - y) dy
$$

The cumulative spectrum is defined as:
$$
\epsilon(k_l) = 1/2<|u_l(x)|^2>,$$
where $k_l=L/l$.
Notice that as filter scale becomes really small ($k_l$ is large), we recover the original signal. While when filter scale becomes very large ($k_l$ approaches 0) we recover the signal mean. [This definition is apparently also available in Uriel Frisch's book on turbulence].

and the Filtering spectrum is:
$$
\overline{E}(k_l) = \frac{d}{d k_l} \epsilon(k_l) 
$$

> Note that this filtering spectrum is capable of obtaining the same power law scaling as the true spectrum, but should not be expected to match the true specturm in an absolute sense. See appendix in Sadek and Aluie 2018 to appreciate this better.
*Probably if we see a match, it is just coincidence or approximate*.
